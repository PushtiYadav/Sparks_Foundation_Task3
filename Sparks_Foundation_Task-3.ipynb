## Category wise analysis of profit,sales,discount

```python
x=df.groupby(by="Category")[['Profit']].sum()
b=x.sort_values(by=['Profit'],ascending=False)
b.plot(kind='bar')
plt.show()
```

```python
df_category = df.groupby(['Category'])[['Sales', 'Discount', 'Profit']].mean()
df_category
```

```python
df_category.plot.pie(subplots=True, figsize=(18, 20), autopct='%1.1f%%', labels = df_category.index, normalize= True)
```

## CONCLUSION:
## Maximun sales and Profit obtain in Technology.
## Minimun profit obtain in Furniture

## Profit distribution according to sales in different regions

```python
plt.figure(figsize=(9,5))
sns.barplot(x='Region',y='Profit',hue='Category',data=df)
plt.show()
```

```python
plt.figure(figsize=(8,8))
df.groupby('Region')['Profit'].sum().plot(kind='pie',autopct='%1.1f%%')
plt.show()
```

## State wise analysis of Profit, Discount, Sale

```python
df_state= df.groupby(['State'])[['Sales', 'Discount', 'Profit']].mean()
df_state.head(10)
```

## Sub-Category wise analysis of profit and sales

## State wise sale analysis 

```python
df_state['Sales'].plot(kind='bar',figsize = (18,5),color='red')
plt.title('State wise analysis of Sale',fontsize=20)
```

## CONCLUSION:

## State wise Profit analysis

```python
df_state[['Profit']].plot(kind = 'bar', figsize = (18,5), color='blue')
plt.title('State wise Profit Analysis', fontsize = 20)
plt.ylabel('Profit per Sate')
plt.xlabel('States')
plt.show()
```

## CONCLUSION:
## Highest Profit-Vermont
## Lowest Profit - Ohio

## State wise discount analysis

```python
df_state['Discount'].plot(kind='bar',figsize=(18,5),color='purple')
plt.title('State wise analysis of Discount', fontsize=20)
```

## CONCLUSION:

## Lowest Discount providing state - lowa,kansas etc(discount=0)
